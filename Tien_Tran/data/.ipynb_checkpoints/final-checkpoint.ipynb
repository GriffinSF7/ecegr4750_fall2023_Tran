```python
import pandas as pd
import numpy as np
from PIL import Image
import os
import re
import torch
import torch.nn as nn
import tqdm
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
import torch.optim as optim
```

```python
# MODEL 1
#I. DATA PREPARATION

#Feature Selection
face_data = pd.read_csv('UTKFaceAugmented.csv')
face_data = face_data.drop(['race', 'remembers_disco', 'uses_skincare'], axis =1)  #drop some unusued features

#convert categorical data to binary
convert_categorical = (['gender', 'age_range', 'has_tiktok'])
for col in convert_categorical:
    face_data = face_data.join(pd.get_dummies(face_data[col], dtype='int'), how = 'outer')  

#convert data to numerical data
convert_numeric = ['age', 'num_haircuts_life', 'max_annual_earnings']
for col in convert_numeric:
    face_data[col] = pd.to_numeric(face_data[col]) 
```

```python
#Process data and train model
features = ['num_haircuts_life', 'max_annual_earnings', 'female', 'yes', 'no', 'male', '0-19', '100-119', '20-29', '30-39',	'40-59', '60-79', '80-99']
target = ['age']
train_data = face_data[features + target]

x_train, x_val, y_train, y_val = train_test_split(train_data[features], train_data[target], train_size=0.7, random_state=42)

x_val, x_test, y_val, y_test = train_test_split(x_val, y_val, train_size=0.6, random_state=42)
print(x_train)

# II. TRAINING MODEL
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test) #make predictions using the test dataset

mse_eval = mean_squared_error(y_test, y_pred) #evaluate the model using MSE metric
print(f'MSE of the model is {mse_eval}')

#since MSE metric yields high error, trying to evaluate model using MAE to reduce outliers 
mae_eval = mean_absolute_error(y_test, y_pred) 
print(f'MAE of the model is {mae_eval}')

#plotting for result visualization
plt.title('Age Prediction')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.scatter(y_test, y_pred, s = 1)
plt.plot([0,100],[0,100], color = 'black', linestyle = '--')
```

```python
def load_dataset(directory):
    images = []
    filenames = []

    for filename in os.listdir(directory):
        img_path = os.path.join(directory, filename)
        img = Image.open(img_path)
        img_array = np.array(img)
        images.append(img_array)
        filenames.append(filename)  # Add the corresponding label for the image
    
    return np.array(images), np.array(filenames)
    
dataset_directory = "/Desktop/Machine Learning/ECEGR4750/ECEGR4750/finals/images"
images, filenames = load_dataset(dataset_directory)
                  
```

```python

#extract age information from filenames
filename_pattern = r'(\d+)\_'  #the first 1 to 3 digits of the filename corresponding to the image's age following by '_'
ages = []
for name in filenames:
    age_pattern = re.search(filename_pattern, name)  #search in the filename for the pattern 
    if age_pattern:
        get_age = age_pattern.group(1) #extract ('d+) to get the age and add to the list
        ages.append(get_age)
ages = np.array(ages)
ages = ages.astype(int)
```

```python

# Split the data into training and testing sets
x1_train, x1_val, y1_train, y1_val = train_test_split(images, ages, train_size=0.7, random_state=42)

# split validation data to get test set
x1_val, x1_test, y1_val, y1_test = train_test_split(x1_val, y1_val, train_size=0.6, random_state=42)


#convert dataframes to tensor format
x1_train = np.array(x1_train)
x1_train = torch.tensor(x1_train)
y1_train = np.array(y1_train)
y1_train = torch.tensor(y1_train)
x1_val = np.array(x1_val)
x1_val = torch.tensor(x1_val)
y1_val = np.array(y1_val)
y1_val = torch.tensor(y1_val)
x1_test = np.array(x1_test)
x1_test = torch.tensor(x1_test)
y1_test = np.array(y1_test)
y1_test = torch.tensor(y1_test)

data_dict = {'x1_train':x1_train, 'y1_train':y1_train, 'x1_val':x1_val, 'y1_val':y1_val, 'x1_test':x1_test, 'y1_test':y1_test}
torch.save(data_dict, './image_age_processed.pt')
```

```python
#Reshape to match the CNN torch input data shape
x1_train = np.reshape(x1_train, (len(x1_train), 3, 200, 200))
x1_val = np.reshape(x1_val, (len(x1_val), 3, 200, 200))
x1_test = np.reshape(x1_test, (len(x1_test), 3, 200, 200))
```

```python
from dataloader import CustomDataloader

train_dataloader = CustomDataloader(x1_train, y1_train, batch_size = 32, randomize=True)
val_dataloader = CustomDataloader(x1_val, y1_val, batch_size = 32)
test_dataloader = CustomDataloader(x1_test, y1_test, batch_size = 32)

#instantiate the model
from networks import DataClassifier

model = DataClassifier(input_dim=3, output_dim=1)

# instantiate Adam optimizer to set learning rate and update weights
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# keep log of losses
epochs = 10  #number of updates for the model
train_losses = []
val_losses = []


loss_fn = torch.nn.BCELoss()  #use cross entropy for loss function
for _ in tqdm.tqdm(range(epochs)):
    #Training model
    t_losses = []
    for _ in range(train_dataloader.num_batches_per_epoch):
        #forward pass
        optimizer.zero_grad()
        train_batch = train_dataloader.fetch_batch()
        prediction = model(train_batch['x_batch'])
        #print(prediction.shape)
        #print(train_batch['y_batch'].shape)
        train_loss = loss_fn(prediction, train_batch['y_batch'].T)
         
        #backward pass
        train_loss.backward()
        optimizer.step()
        t_losses.append(train_loss.detach().numpy())
    train_losses.append(np.mean(t_losses))

    #Validation model
    v_losses = []
    for _ in range(val_dataloader.num_batches_per_epoch):
        #forward pass
        val_batch = val_dataloader.fetch_batch()
        prediction = model(val_batch['x_batch'])
        val_loss = loss_fn(prediction, val_batch['y_batch'])
        v_losses.append(val_loss.detach().numpy())
    val_losses.append(np.mean(v_losses))
```
